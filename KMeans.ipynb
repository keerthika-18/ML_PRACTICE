```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv('C:\\Users\\Keethika P\\Downloads\\archive (4)\\Mall_Customers.csv')
```

```python
df.head()
```

```python
plt.scatter(df['Age'],df['Spending Score (1-100)'],color='r')
```

```python
k_rng=range(1,20)
sse=[]
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(df[['Age','Spending Score (1-100)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(k_rng,sse)
plt.grid()
```

```python
km=KMeans(n_clusters=4)
km
```

```python
y_predict=km.fit_predict(df[['Age','Spending Score (1-100)']])
df['clu']=y_predict
```

```python
df1=df[df.clu==0]
df2=df[df.clu==1]
df3=df[df.clu==2]
df4=df[df.clu==3]
plt.scatter(df1.Age,df1['Spending Score (1-100)'],color='r')
plt.scatter(df2.Age,df2['Spending Score (1-100)'],color='g')
plt.scatter(df3.Age,df3['Spending Score (1-100)'],color='orange')
plt.scatter(df4.Age,df4['Spending Score (1-100)'],color='yellow')
```

```python
scaler=MinMaxScaler()
scaler.fit(df[['Spending Score (1-100)']])
df['Spending Score (1-100)']=scaler.transform(df[['Spending Score (1-100)']])
```

```python
scaler.fit(df[['Age']])
df['Age']=scaler.transform(df[['Age']])
```
